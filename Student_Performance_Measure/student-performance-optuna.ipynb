```python
X=X.drop(['Sleep_Hours','Physical_Activity'],axis=1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
import optuna
from sklearn.model_selection import cross_val_score
import time
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
```

```python
def objective(trial):
    start=time.time()
    classifier_name = trial.suggest_categorical('classifier', ['SVM', 'XGBoost'])

    if classifier_name == 'SVM':
        c = trial.suggest_float('C', 0.1, 100, log=True)
        kernel = trial.suggest_categorical('kernel', ['linear'])
        gamma = trial.suggest_categorical('gamma', ['scale'])
        model = SVR(C=c, kernel=kernel, gamma=gamma)

    elif classifier_name == "XGBoost":
        model = XGBRegressor(
            n_estimators=trial.suggest_int("xgb_n_estimators", 50, 150), 
            learning_rate=trial.suggest_float("xgb_learning_rate", 0.05, 0.3),  
            max_depth=trial.suggest_int("xgb_max_depth", 3, 6),  
            subsample=trial.suggest_float("xgb_subsample", 0.7, 1.0),  
            colsample_bytree=trial.suggest_float("xgb_colsample_bytree", 0.7, 1.0),
            random_state=42,
            tree_method='gpu_hist',
            verbosity=0
        )
    score = cross_val_score(model, X_train, y_train, cv=2, scoring='r2',n_jobs=-1).mean()
    duration = time.time() - start
    if duration > 180:
        print(f"Trial took too long ({duration:.2f} seconds). Pruned.")
        raise optuna.exceptions.TrialPruned()
    return score
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)
```

```python
print("Best trial:")
trial = study.best_trial
print("  R2 Score:", trial.value)
print("  Best Parameters:")
for key, value in trial.params.items():
    print(f"    {key}: {value}")
```

```python
best_params = study.best_trial.params
model = SVR(C=best_params['C'],
        kernel=best_params['kernel'],
        gamma=best_params['gamma']
    )
```

```python
model.fit(X_train,y_train)
ypred=model.predict(X_test)
print(f"r2 score:{r2_score(y_test,ypred)}")
print(f"MSE: {mean_squared_error(y_test,ypred)}")
print(f"MAE: {mean_absolute_error(y_test,ypred)}")
```
